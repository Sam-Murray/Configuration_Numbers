```python
import numpy as np
import itertools as it
```

```python
def dist(couple):
    x = np.array(couple)
    print(x)
    return np.linalg.norm(x[0]-x[1])
    
    
```

```python
odds = list()
evens = list()
x = it.product([0,1], repeat = 4)
for i in x:
    cup = np.array(i)
    if(sum(cup) % 2):
        odds.append(i)
    else:
        evens.append(i)
x = it.product(odds, evens)

mat = list()

for couple in x:
    mat.append(dist(couple))
    
```

```python
mat = np.array(mat)
mat = mat.reshape((8,8))
mat = mat == 1
```
